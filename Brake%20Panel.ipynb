```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
Fs = 8000
f = 5
sample = 8000
x = np.arange(sample)
y = np.sin(2*np.pi*f*x/Fs)
```

```python
plt.plot(x,y)
plt.title('Sinus plot')
plt.xlabel('sample(n)')
plt.ylabel('voltage(V)')
plt.show()
```
